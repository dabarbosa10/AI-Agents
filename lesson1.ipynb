```python
import numpy as np 
```

```python
sentence = "Should we go to a pizzeria or do you prefer a restaurant?"
```

```python
def one_hot(sentences: str):
    vocabulary = sorted(set(sentences.lower().split()))
    word_to_index = dict()
    for i, word in enumerate(vocabulary):
        word_to_index[i] = word
    return word_to_index
```

```python
vocs = one_hot(sentence)
vocs
```
